# Data Loading and Preprocessing

```python
# data loading
location = 'D:/[Analysis]/Data/Sumin/'
raw_main = pd.read_excel(location+'main_KK.xlsx')
raw_main['L11'] = raw_main['L11'].apply(lambda x:1 if x == 1 else 0)
raw_solitude = pd.read_excel(location+'solitude_KK.xlsx')
raw_solitude.columns = ['Region', 'Solitude_2015', 'Solitude_2016', 'Solitude_2017']
raw_aged = pd.read_excel(location+'theaged_KK.xlsx')
raw_aged.columns = ['Region', 'Aged_2014', 'Aged_2015', 'Aged_2016', 'Aged_2017', 'Aged_2018']
raw_region = [[11, '서울특별시'], [21, '부산광역시'], [22, '대구광역시'], [23, '인천광역시'], [24, '광주광역시'], [25, '대전광역시'],
              [26, '울산광역시'], [31, '경기도'], [32, '강원도'], [33, '충청북도'], [34, '충청남도'], [35, '전라북도'], [36, '전라남도'],
              [37, '경상북도'], [38, '경상남도'], [39, '제주특별자치도']]
raw_region = pd.DataFrame(raw_region, columns=['Code', 'Name'])
# display(raw_main.head(), raw_aged.head(), raw_solitude.head(), raw_aged.head(), raw_region.head())

# replace name as code by merging
raw_aged_using = pd.merge(raw_region, raw_aged, how='inner', left_on=['Name'], right_on=['Region'])
raw_solitude_using = pd.merge(raw_region, raw_solitude, how='inner', left_on=['Name'], right_on=['Region'])
# display(raw_aged_using.head(), raw_solitude_using.head())

# delete duplicates
def delete_columns(raw, del_list):
    raw_output = raw.copy()
    for i in del_list:
        if i in raw_output.columns:
            del raw_output[i]
    return raw_output
raw_aged_using = delete_columns(raw_aged_using, ['Name', 'Region'])
raw_solitude_using = delete_columns(raw_solitude_using, ['Name', 'Region'])
# display(raw_aged_using.head(), raw_solitude_using.head())

# join aged and solitude to main raw
if 'Aged_2014' not in raw_main.columns:
    raw_main_using = pd.merge(raw_main, raw_aged_using, how='inner', left_on=['id1_1'], right_on=['Code'])
if 'Solitude_2015' not in raw_main_using.columns:
    raw_main_using = pd.merge(raw_main_using, raw_solitude_using, how='inner', left_on=['id1_1'], right_on=['Code'])

# removing 9 in L11
raw_main_using = raw_main_using[raw_main_using['L11'] != 9]

# descriptive statistics
raw_main_using.rename(columns={'Code_x':'Region'}, inplace=True)
raw_main_using = delete_columns(raw_main_using, ['id1_1', 'Aged_2014', 'Aged_2016', 'Aged_2017', 'Aged_2018', 'Solitude_2016', 'Solitude_2017', 'Code_y'])
display(raw_main_using.head(), raw_main_using.describe().T)
```

# Analysis (Regression)

```python
Y_column = ['L11']
X_columns_level2 = ['Region', 'Aged_2015', 'Solitude_2015']
X_columns_level1 = [col for col in raw_main_using.columns if col not in Y_column+X_columns_level2]
raw_Y = raw_main_using[Y_column]
```

```python
raw_X = raw_main_using[X_columns_level1]

logistic_result = sm.Logit(raw_Y, raw_X).fit()
display(logistic_result.summary(), np.exp(logistic_result.params))
```

```python
raw_X = pd.concat([raw_main_using[X_columns_level1], 
                   pd.get_dummies(raw_main_using['Region'], prefix='Region_')], axis=1)

logistic_result = sm.Logit(raw_Y, raw_X).fit()
display(logistic_result.summary(), np.exp(logistic_result.params))
```

```python
raw_X = raw_main_using[X_columns_level1+['Aged_2015']]

logistic_result = sm.Logit(raw_Y, raw_X).fit()
display(logistic_result.summary(), np.exp(logistic_result.params))
```

```python
raw_X = raw_main_using[X_columns_level1+['Solitude_2015']]

logistic_result = sm.Logit(raw_Y, raw_X).fit()
display(logistic_result.summary(), np.exp(logistic_result.params))
```

```python
raw_X = raw_main_using[X_columns_level1]
raw_X_level2 = raw_main_using['Region']

formula = ' ~ '.join([''.join(raw_Y.columns), ' + '.join(raw_X.columns)])
mlm_result = smf.mixedlm(formula, raw_main_using, groups=raw_X_level2, use_sparse=True).fit()
display(mlm_result.summary(), np.exp(mlm_result.params))
```

```python
raw_X = raw_main_using[X_columns_level1]
raw_X_level2 = raw_main_using['Aged_2015']

formula = ' ~ '.join([''.join(raw_Y.columns), ' + '.join(raw_X.columns)])
mlm_result = smf.mixedlm(formula, raw_main_using, groups=raw_X_level2, use_sparse=True).fit()
display(mlm_result.summary(), np.exp(mlm_result.params))
```

```python
raw_X = raw_main_using[X_columns_level1]
raw_X_level2 = raw_main_using['Solitude_2015']

formula = ' ~ '.join([''.join(raw_Y.columns), ' + '.join(raw_X.columns)])
mlm_result = smf.mixedlm(formula, raw_main_using, groups=raw_X_level2, use_sparse=True).fit()
display(mlm_result.summary(), np.exp(mlm_result.params))
```

# Analysis (Data Mining)

```python
raw_X = raw_main_using[X_columns_level1]

rf_result = RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
                       max_features='auto', min_samples_split=2,
                       n_estimators=100, n_jobs=2, random_state=123).fit(raw_X, raw_Y)
rf_result_import = pd.DataFrame([rf_result.feature_importances_], 
                              columns=raw_X.columns, 
                              index=['importance']).T.sort_values(by=['importance'], ascending=False)
rf_result_import.plot.bar(figsize=(12,6), fontsize=15)
plt.title('Variable Importances', fontsize=15)
plt.show()
```

```python
raw_X = pd.concat([raw_main_using[X_columns_level1], 
                   pd.get_dummies(raw_main_using['Region'], prefix='Region_')], axis=1)

rf_result = RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
                       max_features='auto', min_samples_split=2,
                       n_estimators=100, n_jobs=2, random_state=123).fit(raw_X, raw_Y)
rf_result_import = pd.DataFrame([rf_result.feature_importances_], 
                              columns=raw_X.columns, 
                              index=['importance']).T.sort_values(by=['importance'], ascending=False)
rf_result_import.plot.bar(figsize=(12,6), fontsize=15)
plt.title('Variable Importances', fontsize=15)
plt.show()
```
