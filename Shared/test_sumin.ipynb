```python
# data loading
location = 'D:/[Analysis]/Data/Sumin/'
raw_main = pd.read_excel(location+'main_KK.xlsx')
raw_solitude = pd.read_excel(location+'solitude_KK.xlsx')
raw_solitude.columns = ['Region', 'Solitude_2015', 'Solitude_2016', 'Solitude_2017']
raw_aged = pd.read_excel(location+'theaged_KK.xlsx')
raw_aged.columns = ['Region', 'Aged_2014', 'Aged_2015', 'Aged_2016', 'Aged_2017', 'Aged_2018']
raw_region = [[11, '서울특별시'], [21, '부산광역시'], [22, '대구광역시'], [23, '인천광역시'], [24, '광주광역시'], [25, '대전광역시'],
              [26, '울산광역시'], [31, '경기도'], [32, '강원도'], [33, '충청북도'], [34, '충청남도'], [35, '전라북도'], [36, '전라남도'],
              [37, '경상북도'], [38, '경상남도'], [39, '제주특별자치도']]
raw_region = pd.DataFrame(raw_region, columns=['Code', 'Name'])
display(raw_main.head(), raw_aged.head(), raw_solitude.head(), raw_aged.head(), raw_region.head())
```

```python
# replace name as code by merging
raw_aged_using = pd.merge(raw_region, raw_aged, how='inner', left_on=['Name'], right_on=['Region'])
raw_solitude_using = pd.merge(raw_region, raw_solitude, how='inner', left_on=['Name'], right_on=['Region'])
display(raw_aged_using.head(), raw_solitude_using.head())

# delete duplicates
def delete_columns(raw, del_list):
    raw_output = raw.copy()
    for i in del_list:
        if i in raw_output.columns:
            del raw_output[i]
    return raw_output
raw_aged_using = delete_columns(raw_aged_using, ['Name', 'Region'])
raw_solitude_using = delete_columns(raw_solitude_using, ['Name', 'Region'])
display(raw_aged_using.head(), raw_solitude_using.head())

# join aged and solitude to main raw
if 'Aged_2014' not in raw_main.columns:
    raw_main_using = pd.merge(raw_main, raw_aged_using, how='inner', left_on=['id1_1'], right_on=['Code'])
if 'Solitude_2015' not in raw_main_using.columns:
    raw_main_using = pd.merge(raw_main_using, raw_solitude_using, how='inner', left_on=['id1_1'], right_on=['Code'])

# removing 9 in L11
raw_main_using = raw_main_using[raw_main_using['L11'] != 9]

# descriptive statistics
raw_main_using = delete_columns(raw_main_using, ['Aged_2014', 'Aged_2016', 'Aged_2017', 'Aged_2018', 'Solitude_2016', 'Solitude_2017', 'Code_y'])
raw_main_using.rename(columns={'Code_x':'Region'}, inplace=True)
display(raw_main_using.head())
raw_main_using.describe().T
```
